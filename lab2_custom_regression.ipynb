# Линейная регрессия с исп градиентного спуска

- y = b0 + b1X1+...bnXn,

где Xi — независимая переменная, а y — зависимая переменная, коэффициенты регрессии (наклон для каждой независимой переменной) = b1,...bn, а b0 — точка пересечения с осью y.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
def linear_regression(X, y, learning_rate=0.01, n_iters=1000, alpha = 0.1):
    n_samples, n_features = X.shape
    
    weights = np.zeros(n_features)
    bias = 0
    
    # Градиентный спуск
    for _ in range(n_iters):
        y_pred = np.dot(X, weights) + bias
        
        # Градиенты
        error = y_pred - y
        dw = (2 / n_samples) * np.dot(X.T, error) + 2 * alpha * weights
        db = (2 / n_samples) * np.sum(error)
        
        weights -= learning_rate * dw
        bias -= learning_rate * db
    
    return weights, bias

def predict(X, weights, bias):
    return np.dot(X, weights) + bias
```
