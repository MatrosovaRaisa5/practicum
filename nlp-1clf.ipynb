### Предикторы
- Date - дата проведения наблюдений
- Location - название метеостанции/географическое местоположение, где были собраны данные
- MinTemp - минимальная температура за день (°C)
- MaxTemp - максимальная температура за день (°C)
- Temp9am - температура в 9:00 утра (°C)
- Temp3pm - температура в 15:00 дня (°C)
- Rainfall - количество выпавших осадков за день (мм)
- RainToday - бинарный показатель: были ли осадки сегодня (Yes/No)
- WindGustDir - направление самого сильного порыва ветра за день
- WindGustSpeed - скорость самого сильного порыва ветра (км/ч)
- WindDir9am - преобладающее направление ветра в 9:00
- WindDir3pm - преобладающее направление ветра в 15:00
- WindSpeed9am - скорость ветра в 9:00 (км/ч)
- WindSpeed3pm - скорость ветра в 15:00 (км/ч)
- Humidity9am - относительная влажность в 9:00 (%)
- Humidity3pm - относительная влажность в 15:00 (%)
- Cloud9am - облачность в 9:00 (в восьмых долях: 0 = ясно, 8 = полностью облачно)
- Cloud3pm - облачность в 15:00 (в восьмых долях)
- Pressure9am - атмосферное давление в 9:00 (гПа)
- Pressure3pm - атмосферное давление в 15:00 (гПа)
- Evaporation - величина испарения за 24 часа (мм)
- Sunshine - продолжительность солнечного сияния за день (часы)
- RainTomorrow - целевая переменная для прогнозирования. Она отвечает на вопрос: будет ли дождь на следующий день? (Да или Нет). В этом столбце ставится отметка «Да», если в этот день выпало 1 мм осадков или больше.

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
import torch
from torch.utils.data import Dataset, DataLoader
```

```python
data_processed = data.copy()
wind_columns = ['WindGustDir', 'WindDir9am', 'WindDir3pm']
le_dict = {}

for col in wind_columns:
    le = LabelEncoder()
    data_processed[col] = le.fit_transform(data_processed[col])
    le_dict[col] = le

data_processed = pd.get_dummies(data_processed, columns=['Location'], prefix='loc')
data_processed
```

```python
label = 'RainTomorrow'

features = [col for col in data_processed.columns if col != label]

X = data_processed[features].values
y = data_processed[label].values
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
y = y.astype(int)
print(f"До балансировки: {Counter(y)}")

smote = SMOTE(random_state=42)
X_balanced, y_balanced = smote.fit_resample(X, y)

print(f"Балансировка: {Counter(y_balanced)}")
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
k_best = SelectKBest(score_func=f_classif, k=30)
X_final = k_best.fit_transform(X_balanced, y_balanced) 

print(f"Финальное количество признаков: {X_final.shape[1]}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_final, y_balanced, test_size=0.2, random_state=42, stratify=y_balanced)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_tensor = torch.FloatTensor(X_train_scaled)
y_train_tensor = torch.LongTensor(y_train)
X_test_tensor = torch.FloatTensor(X_test_scaled)
y_test_tensor = torch.LongTensor(y_test)

class WeatherDataset(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels
        
    def __len__(self):
        return len(self.features)
    
    def __getitem__(self, idx):
        return self.features[idx], self.labels[idx] 
```

```python
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report, roc_auc_score, roc_curve

class ClassificationModel(nn.Module):
    def __init__(self, input_size, num_classes):
        super(ClassificationModel, self).__init__()
        self.layer1 = nn.Linear(input_size, 256)
        self.relu1 = nn.ReLU()
        self.dropout1 = nn.Dropout(0.1)
        
        self.layer2 = nn.Linear(256, 128)
        self.relu2 = nn.ReLU()
        self.dropout2 = nn.Dropout(0.1)
        
        self.layer3 = nn.Linear(128, 64)
        self.relu3 = nn.ReLU()
    
        self.output = nn.Linear(64, num_classes)
    
    def forward(self, x):
        x = self.dropout1(self.relu1(self.layer1(x)))
        x = self.dropout2(self.relu2(self.layer2(x)))
        x = self.relu3(self.layer3(x))
        x = self.output(x)
        return x
    
    def predict_proba(self, x):
        with torch.no_grad():
            logits = self.forward(x)
            return torch.softmax(logits, dim=1)
    
    def predict(self, x):
        with torch.no_grad():
            probabilities = self.predict_proba(x)
            return torch.argmax(probabilities, dim=1)
```

```python
num_classes = len(np.unique(y))
y_train_tensor = torch.LongTensor(y_train) 
y_test_tensor = torch.LongTensor(y_test)

print(f"Количество классов: {num_classes}")
print(f"Размер y_train_tensor: {y_train_tensor.shape}")
print(f"Размер y_test_tensor: {y_test_tensor.shape}")

train_dataset = WeatherDataset(X_train_tensor, y_train_tensor)
test_dataset = WeatherDataset(X_test_tensor, y_test_tensor)

train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=128, shuffle=False)

input_size = X_train_tensor.shape[1]
model = ClassificationModel(input_size, num_classes)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.0005)

train_losses = []
val_losses = []

epochs = 50

for epoch in range(epochs):
    model.train()
    epoch_train_loss = 0
    batch_count = 0
    
    for x_b, y_b in train_loader:
        outputs = model(x_b)
        
        loss_value = criterion(outputs, y_b)
        loss_value.backward()
        optimizer.step()
        optimizer.zero_grad()
        
        epoch_train_loss += loss_value.item()
        batch_count += 1
    
    avg_train_loss = epoch_train_loss / batch_count
    train_losses.append(avg_train_loss)

    model.eval()
    epoch_val_loss = 0
    batch_count_val = 0
    
    with torch.no_grad():
        for x_b, y_b in test_loader:
            outputs = model(x_b)
            loss_value = criterion(outputs, y_b)
            epoch_val_loss += loss_value.item()
            batch_count_val += 1
    
    avg_val_loss = epoch_val_loss / batch_count_val
    val_losses.append(avg_val_loss)
    
    print(f'Эпоха {epoch + 1}, Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}')
```

```python
plt.figure(figsize=(12, 6))
plt.plot(range(1, epochs + 1), train_losses, label='Train Loss', marker='o')
plt.plot(range(1, epochs + 1), val_losses, label='Val Loss', marker='o')
plt.xlabel('Эпохи')
plt.ylabel('CrossEntropy Loss')
plt.title('График функции ошибки по эпохам')
plt.legend()
plt.grid(True)
plt.xticks(range(0, epochs + 1, 10))
plt.show()

model.eval()
all_predictions = []
all_targets = []

with torch.no_grad():
    for x_b, y_b in test_loader:
        outputs = model(x_b)
        predictions = torch.argmax(outputs, dim=1)
        
        all_predictions.extend(predictions.cpu().detach().numpy())
        all_targets.extend(y_b.cpu().detach().numpy())

y_pred = np.array(all_predictions)
y_true = np.array(all_targets)

accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='weighted', zero_division=0)
recall = recall_score(y_true, y_pred, average='weighted', zero_division=0)
f1 = f1_score(y_true, y_pred, average='weighted', zero_division=0)

print("ОЦЕНКА КАЧЕСТВА МОДЕЛИ")
print(f"Accuracy (Точность): {accuracy:.4f}")
print(f"Precision (Точность, weighted): {precision:.4f}")
print(f"Recall (Полнота, weighted): {recall:.4f}")
print(f"F1-Score (weighted): {f1:.4f}")
print("\nClassification Report:")
print(classification_report(y_true, y_pred, zero_division=0))
```
